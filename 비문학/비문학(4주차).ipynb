## PDF에서 비문학 문제만 추출(이미지로 분할 후 저장)

```python
import os
from pathlib import Path
from PyPDF2 import PdfReader
import ksatparser

# 대상 파일 경로
pdf_path = "/content/drive/MyDrive/기출/2026-06-국어.pdf"

# 텍스트 저장 경로
text_save_dir = "/content/drive/MyDrive/save_text"
os.makedirs(text_save_dir, exist_ok=True)
text_save_path = os.path.join(text_save_dir, "2026-06-국어.txt")

# 이미지 저장 경로
image_output_dir = "/content/drive/MyDrive/output_images"
os.makedirs(image_output_dir, exist_ok=True)

# 텍스트 추출 함수 정의
def extract_text_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    full_text = ""
    for page in reader.pages:
        text = page.extract_text()
        if text:
            full_text += text + "\n"
    return full_text

# 파일 처리
if os.path.exists(pdf_path):
    # 텍스트 추출 및 저장
    raw_text = extract_text_from_pdf(pdf_path)
    with open(text_save_path, 'w', encoding='utf-8') as f:
        f.write(raw_text)
    print(f"텍스트 저장 완료: {text_save_path}")

    # 이미지 추출
    ksatparser.parse_problem(pdf_path, image_output_dir)
    print(f"이미지 추출 완료: {image_output_dir}")
else:
    print(f"파일이 존재하지 않습니다: {pdf_path}")
```

## LLM 성능평가 (텍스트)

```python
import os
from openai import OpenAI
import re
import time

client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])

# 📌 프롬프트 템플릿
text_prompt = '''
다음은 국어 모의고사 비문학 문제입니다. 지문과 문제를 읽고 아래를 수행하세요:

1. 이 문제의 정답을 알려주세요 (선지 번호만 출력: ①~⑤ 중 하나)
2. 반드시 지문 내용을 근거로 하여 정답 선택 이유를 자세히 설명해주세요 (3~5문장)

[지문]
{passage}

[문제]
{question}

결과는 다음 형식으로 출력해주세요:

[정답] ③
[해설] … (여기에 근거 설명)
'''

# ✅ 지문-문제 분리 함수
def split_passages_and_questions(text):
    # 지문 추출: [1~3], [4~9] 등 패턴
    passage_pattern = re.compile(r"\[(\d+)[~～]?\d*\] ?(.+?)(?=\[\d+|1\.|\Z)", re.DOTALL)
    passages = passage_pattern.findall(text)

    # 지문 번호로 dict 생성
    passage_dict = {}
    for num, passage in passages:
        passage_dict[int(num)] = passage.strip()

    # 문제 추출: 1.문제 + ① ~ ⑤
    # Modified regex to avoid variable-width look-behind
    question_pattern = re.compile(
        r"(?:^|\n)(\d{1,2})\.(.+?)(?=\n\d{1,2}\.|$)", re.DOTALL
    )
    questions = question_pattern.findall(text)

    # 문제별 지문 매칭
    paired_data = []
    for qnum_str, question_block in questions:
        qnum = int(qnum_str)
        # 가장 가까운 지문 번호 찾기
        passage_key = max([k for k in passage_dict if k <= qnum], default=1)
        passage = passage_dict[passage_key]
        question = question_block.strip()
        paired_data.append((qnum, passage, question))

    return paired_data

# ✅ GPT 요청 함수
def get_answer_explanation(passage, question):
    prompt = text_prompt.format(passage=passage, question=question)
    try:
        response = client.chat.completions.create(
            model="gpt-4o",
            messages=[
                {"role": "system", "content": "너는 국어 비문학 문제 해설 전문가야."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.2
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        return f"[Error]: {e}"

# ✅ 전체 실행
def run_full_evaluation(text_path):
    with open(text_path, encoding="utf-8") as f:
        raw_text = f.read()

    parsed = split_passages_and_questions(raw_text)
    results = []

    for qnum, passage, question in parsed:
        print(f"\n문제 {qnum}")
        result = get_answer_explanation(passage, question)
        print(result)
        print("-" * 80)
        results.append((qnum, result))
        time.sleep(1.5)  # (선택) rate limit 보호

    return results

# ✅ 실행
results = run_full_evaluation("/content/2026-06-국어.txt")
```

```python
import os
import re
import unicodedata
import easyocr
from PIL import Image
from openai import OpenAI

# 유니코드 NFC 정규화
def normalize_filename(fn):
    return unicodedata.normalize('NFC', fn)

# EasyOCR 리더 초기화 (한글 + 숫자, 특수기호 혼합 언어)
reader = easyocr.Reader(['ko', 'en'], gpu=False)

# OCR 함수: 이미지 → 텍스트 (후처리 포함)
def extract_text(image_path):
    results = reader.readtext(image_path, detail=0, paragraph=True)
    text = '\n'.join(results)
    # 보기 번호 매핑 보정: "1." → "①", "2)" → "②" 등
    text = re.sub(r'\b1[\.\)]', '①', text)
    text = re.sub(r'\b2[\.\)]', '②', text)
    text = re.sub(r'\b3[\.\)]', '③', text)
    text = re.sub(r'\b4[\.\)]', '④', text)
    text = re.sub(r'\b5[\.\)]', '⑤', text)
    return text

# GPT 호출 함수 (변경 없음)
def ask_gpt(prompt_text):
    client = OpenAI()
    resp = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role":"user","content":prompt_text}],
        temperature=0.3
    )
    return resp.choices[0].message.content

# 프롬프트 템플릿
text_prompt = '''
다음은 국어 모의고사 비문학 문제입니다. 지문과 문제를 읽고 아래를 수행하세요:

1. 이 문제의 정답을 알려주세요 (선지 번호만 출력: ①~⑤ 중 하나)
2. 반드시 지문 내용을 근거로 하여 정답 선택 이유를 자세히 설명해주세요 (3~5문장)

[지문]
{passage}

[문제]
{question}

결과는 다음 형식으로 출력해주세요:

[정답] ③
[해설] … (여기에 근거 설명)
'''

# 이미지-문제 mapping
mapping = {
    "p1": [1, 2, 3],
    "p2": list(range(4, 10)),
    "p3": list(range(10, 14)),
    "p4": list(range(14, 18))
}

image_dir = "/content/drive/MyDrive/classified_images/비문학"
all_files = [normalize_filename(f) for f in os.listdir(image_dir)]

for p_key, q_nums in mapping.items():
    # 지문
    pf = normalize_filename(f"2026-06-국어_{p_key}.png")
    if pf not in all_files:
        print(f"❌ 지문 없음: {pf}")
        continue
    passage = extract_text(os.path.join(image_dir, pf)).strip()

    for qn in q_nums:
        qf = normalize_filename(f"2026-06-국어_{qn}.png")
        if qf not in all_files:
            print(f"❌ 문제 없음: {qf}")
            continue
        question = extract_text(os.path.join(image_dir, qf)).strip()

        prompt = text_prompt.format(passage=passage, question=question)
        print(f"\n📘 [문제 {qn}]")
        try:
            print(ask_gpt(prompt))
        except Exception as e:
            print(f"⚠️ GPT 실패: {e}")
```

```python
import os
import re
import cv2
import numpy as np
import unicodedata
import easyocr
from PIL import Image
from openai import OpenAI

# ✅ 유니코드 NFC 정규화
def normalize_filename(fn):
    return unicodedata.normalize('NFC', fn)

# ✅ EasyOCR 초기화
reader = easyocr.Reader(['ko', 'en'], gpu=False)

# ✅ 텍스트 추출 + 밑줄 강조 영역 감지 함수
def extract_text_with_underlines(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # EasyOCR 결과: (bbox, text, confidence)
    results = reader.readtext(gray, detail=True)

    # Canny + Hough로 밑줄 선 검출
    edges = cv2.Canny(gray, 50, 150, apertureSize=3)
    lines = cv2.HoughLinesP(edges, 1, np.pi/180, threshold=100,
                            minLineLength=50, maxLineGap=10)

    underlined_texts = set()
    if lines is not None:
        for (bbox, text, conf) in results:
            # 텍스트의 y중심 좌표
            text_y = np.mean([point[1] for point in bbox])
            for line in lines:
                x1, y1, x2, y2 = line[0]
                line_y = (y1 + y2) / 2
                if abs(line_y - text_y) < 10:
                    underlined_texts.add(text)
                    break

    # 전체 텍스트 재구성
    full_text = ""
    for (bbox, text, conf) in results:
        if text in underlined_texts:
            full_text += f"**{text}** "
        else:
            full_text += f"{text} "

    # 보기 번호 매핑
    full_text = re.sub(r'\b1[\.\)]', '①', full_text)
    full_text = re.sub(r'\b2[\.\)]', '②', full_text)
    full_text = re.sub(r'\b3[\.\)]', '③', full_text)
    full_text = re.sub(r'\b4[\.\)]', '④', full_text)
    full_text = re.sub(r'\b5[\.\)]', '⑤', full_text)

    return full_text.strip()

# ✅ GPT 호출 함수
def ask_gpt(prompt_text):
    client = OpenAI()
    resp = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role":"user","content":prompt_text}],
        temperature=0.3
    )
    return resp.choices[0].message.content

# ✅ GPT 프롬프트
text_prompt = '''
다음은 국어 모의고사 비문학 문제입니다. 지문과 문제를 읽고 아래를 수행하세요:

1. 이 문제의 정답을 알려주세요 (선지 번호만 출력: ①~⑤ 중 하나)
2. 반드시 지문 내용을 근거로 하여 정답 선택 이유를 자세히 설명해주세요 (3~5문장)

[지문]
{passage}

[문제]
{question}

결과는 다음 형식으로 출력해주세요:

[정답] ③
[해설] … (여기에 근거 설명)
'''

# ✅ 이미지-문제 매핑
mapping = {
    "p1": [1, 2, 3],
    "p2": list(range(4, 10)),
    "p3": list(range(10, 14)),
    "p4": list(range(14, 18))
}

# ✅ 실행
image_dir = "/content/drive/MyDrive/classified_images/비문학"
all_files = [normalize_filename(f) for f in os.listdir(image_dir)]

for p_key, q_nums in mapping.items():
    pf = normalize_filename(f"2026-06-국어_{p_key}.png")
    passage = extract_text_with_underlines(os.path.join(image_dir, pf))

    for qn in q_nums:
        qf = normalize_filename(f"2026-06-국어_{qn}.png")
        question = extract_text_with_underlines(os.path.join(image_dir, qf))

        prompt = text_prompt.format(passage=passage, question=question)
        print(f"\n📘 [문제 {qn}]")
        try:
            print(ask_gpt(prompt))
        except Exception as e:
            print(f"⚠️ GPT 실패: {e}")
```

**밑줄, <보기> 박스 강조**

```python
import cv2
import numpy as np
import os

def remove_underlines(input_path, output_path):
    # 이미지 불러오기
    img = cv2.imread(input_path)
    if img is None:
        raise FileNotFoundError(f"❌ 이미지 경로 확인: {input_path}")

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # 이진화
    _, binary = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY_INV)

    # 수평선 검출용 커널 (1x30)
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (30, 1))
    detected_lines = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel, iterations=2)

    # 수평선 영역 제거
    cleaned = cv2.inpaint(img, detected_lines, inpaintRadius=3, flags=cv2.INPAINT_TELEA)

    # 저장
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    cv2.imwrite(output_path, cleaned)
    print(f"✅ 밑줄 제거 이미지 저장됨: {output_path}")
```
