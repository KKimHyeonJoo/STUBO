## LLM ÏÑ±Îä•ÌèâÍ∞Ä (ÌÖçÏä§Ìä∏)

```python
import os
from openai import OpenAI
import re
import time

client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])

# üìå ÌîÑÎ°¨ÌîÑÌä∏ ÌÖúÌîåÎ¶ø
text_prompt = '''
Îã§ÏùåÏùÄ Íµ≠Ïñ¥ Î™®ÏùòÍ≥†ÏÇ¨ ÎπÑÎ¨∏Ìïô Î¨∏Ï†úÏûÖÎãàÎã§. ÏßÄÎ¨∏Í≥º Î¨∏Ï†úÎ•º ÏùΩÍ≥† ÏïÑÎûòÎ•º ÏàòÌñâÌïòÏÑ∏Ïöî:

1. Ïù¥ Î¨∏Ï†úÏùò Ï†ïÎãµÏùÑ ÏïåÎ†§Ï£ºÏÑ∏Ïöî (ÏÑ†ÏßÄ Î≤àÌò∏Îßå Ï∂úÎ†•: ‚ë†~‚ë§ Ï§ë ÌïòÎÇò)
2. Î∞òÎìúÏãú ÏßÄÎ¨∏ ÎÇ¥Ïö©ÏùÑ Í∑ºÍ±∞Î°ú ÌïòÏó¨ Ï†ïÎãµ ÏÑ†ÌÉù Ïù¥Ïú†Î•º ÏûêÏÑ∏Ìûà ÏÑ§Î™ÖÌï¥Ï£ºÏÑ∏Ïöî (3~5Î¨∏Ïû•)

[ÏßÄÎ¨∏]
{passage}

[Î¨∏Ï†ú]
{question}

Í≤∞Í≥ºÎäî Îã§Ïùå ÌòïÏãùÏúºÎ°ú Ï∂úÎ†•Ìï¥Ï£ºÏÑ∏Ïöî:

[Ï†ïÎãµ] ‚ë¢
[Ìï¥ÏÑ§] ‚Ä¶ (Ïó¨Í∏∞Ïóê Í∑ºÍ±∞ ÏÑ§Î™Ö)
'''

# ‚úÖ ÏßÄÎ¨∏-Î¨∏Ï†ú Î∂ÑÎ¶¨ Ìï®Ïàò
def split_passages_and_questions(text):
    # ÏßÄÎ¨∏ Ï∂îÏ∂ú: [1~3], [4~9] Îì± Ìå®ÌÑ¥
    passage_pattern = re.compile(r"\[(\d+)[~ÔΩû]?\d*\] ?(.+?)(?=\[\d+|1\.|\Z)", re.DOTALL)
    passages = passage_pattern.findall(text)

    # ÏßÄÎ¨∏ Î≤àÌò∏Î°ú dict ÏÉùÏÑ±
    passage_dict = {}
    for num, passage in passages:
        passage_dict[int(num)] = passage.strip()

    # Î¨∏Ï†ú Ï∂îÏ∂ú: 1.Î¨∏Ï†ú + ‚ë† ~ ‚ë§
    # Modified regex to avoid variable-width look-behind
    question_pattern = re.compile(
        r"(?:^|\n)(\d{1,2})\.(.+?)(?=\n\d{1,2}\.|$)", re.DOTALL
    )
    questions = question_pattern.findall(text)

    # Î¨∏Ï†úÎ≥Ñ ÏßÄÎ¨∏ Îß§Ïπ≠
    paired_data = []
    for qnum_str, question_block in questions:
        qnum = int(qnum_str)
        # Í∞ÄÏû• Í∞ÄÍπåÏö¥ ÏßÄÎ¨∏ Î≤àÌò∏ Ï∞æÍ∏∞
        passage_key = max([k for k in passage_dict if k <= qnum], default=1)
        passage = passage_dict[passage_key]
        question = question_block.strip()
        paired_data.append((qnum, passage, question))

    return paired_data

# ‚úÖ GPT ÏöîÏ≤≠ Ìï®Ïàò
def get_answer_explanation(passage, question):
    prompt = text_prompt.format(passage=passage, question=question)
    try:
        response = client.chat.completions.create(
            model="gpt-4o",
            messages=[
                {"role": "system", "content": "ÎÑàÎäî Íµ≠Ïñ¥ ÎπÑÎ¨∏Ìïô Î¨∏Ï†ú Ìï¥ÏÑ§ Ï†ÑÎ¨∏Í∞ÄÏïº."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.2
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        return f"[Error]: {e}"

# ‚úÖ Ï†ÑÏ≤¥ Ïã§Ìñâ
def run_full_evaluation(text_path):
    with open(text_path, encoding="utf-8") as f:
        raw_text = f.read()

    parsed = split_passages_and_questions(raw_text)
    results = []

    for qnum, passage, question in parsed:
        print(f"\nÎ¨∏Ï†ú {qnum}")
        result = get_answer_explanation(passage, question)
        print(result)
        print("-" * 80)
        results.append((qnum, result))
        time.sleep(1.5)  # (ÏÑ†ÌÉù) rate limit Î≥¥Ìò∏

    return results

# ‚úÖ Ïã§Ìñâ
results = run_full_evaluation("/content/2026-06-·ÑÄ·ÖÆ·Ü®·Ñã·Ö•.txt")
```

```python
import os
import re
import unicodedata
import easyocr
from PIL import Image
from openai import OpenAI

# Ïú†ÎãàÏΩîÎìú NFC Ï†ïÍ∑úÌôî
def normalize_filename(fn):
    return unicodedata.normalize('NFC', fn)

# EasyOCR Î¶¨Îçî Ï¥àÍ∏∞Ìôî (ÌïúÍ∏Ä + Ïà´Ïûê, ÌäπÏàòÍ∏∞Ìò∏ ÌòºÌï© Ïñ∏Ïñ¥)
reader = easyocr.Reader(['ko', 'en'], gpu=False)

# OCR Ìï®Ïàò: Ïù¥ÎØ∏ÏßÄ ‚Üí ÌÖçÏä§Ìä∏ (ÌõÑÏ≤òÎ¶¨ Ìè¨Ìï®)
def extract_text(image_path):
    results = reader.readtext(image_path, detail=0, paragraph=True)
    text = '\n'.join(results)
    # Î≥¥Í∏∞ Î≤àÌò∏ Îß§Ìïë Î≥¥Ï†ï: "1." ‚Üí "‚ë†", "2)" ‚Üí "‚ë°" Îì±
    text = re.sub(r'\b1[\.\)]', '‚ë†', text)
    text = re.sub(r'\b2[\.\)]', '‚ë°', text)
    text = re.sub(r'\b3[\.\)]', '‚ë¢', text)
    text = re.sub(r'\b4[\.\)]', '‚ë£', text)
    text = re.sub(r'\b5[\.\)]', '‚ë§', text)
    return text

# GPT Ìò∏Ï∂ú Ìï®Ïàò (Î≥ÄÍ≤Ω ÏóÜÏùå)
def ask_gpt(prompt_text):
    client = OpenAI()
    resp = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role":"user","content":prompt_text}],
        temperature=0.3
    )
    return resp.choices[0].message.content

# ÌîÑÎ°¨ÌîÑÌä∏ ÌÖúÌîåÎ¶ø
text_prompt = '''
Îã§ÏùåÏùÄ Íµ≠Ïñ¥ Î™®ÏùòÍ≥†ÏÇ¨ ÎπÑÎ¨∏Ìïô Î¨∏Ï†úÏûÖÎãàÎã§. ÏßÄÎ¨∏Í≥º Î¨∏Ï†úÎ•º ÏùΩÍ≥† ÏïÑÎûòÎ•º ÏàòÌñâÌïòÏÑ∏Ïöî:

1. Ïù¥ Î¨∏Ï†úÏùò Ï†ïÎãµÏùÑ ÏïåÎ†§Ï£ºÏÑ∏Ïöî (ÏÑ†ÏßÄ Î≤àÌò∏Îßå Ï∂úÎ†•: ‚ë†~‚ë§ Ï§ë ÌïòÎÇò)
2. Î∞òÎìúÏãú ÏßÄÎ¨∏ ÎÇ¥Ïö©ÏùÑ Í∑ºÍ±∞Î°ú ÌïòÏó¨ Ï†ïÎãµ ÏÑ†ÌÉù Ïù¥Ïú†Î•º ÏûêÏÑ∏Ìûà ÏÑ§Î™ÖÌï¥Ï£ºÏÑ∏Ïöî (3~5Î¨∏Ïû•)

[ÏßÄÎ¨∏]
{passage}

[Î¨∏Ï†ú]
{question}

Í≤∞Í≥ºÎäî Îã§Ïùå ÌòïÏãùÏúºÎ°ú Ï∂úÎ†•Ìï¥Ï£ºÏÑ∏Ïöî:

[Ï†ïÎãµ] ‚ë¢
[Ìï¥ÏÑ§] ‚Ä¶ (Ïó¨Í∏∞Ïóê Í∑ºÍ±∞ ÏÑ§Î™Ö)
'''

# Ïù¥ÎØ∏ÏßÄ-Î¨∏Ï†ú mapping
mapping = {
    "p1": [1, 2, 3],
    "p2": list(range(4, 10)),
    "p3": list(range(10, 14)),
    "p4": list(range(14, 18))
}

image_dir = "/content/drive/MyDrive/classified_images/ÎπÑÎ¨∏Ìïô"
all_files = [normalize_filename(f) for f in os.listdir(image_dir)]

for p_key, q_nums in mapping.items():
    # ÏßÄÎ¨∏
    pf = normalize_filename(f"2026-06-Íµ≠Ïñ¥_{p_key}.png")
    if pf not in all_files:
        print(f"‚ùå ÏßÄÎ¨∏ ÏóÜÏùå: {pf}")
        continue
    passage = extract_text(os.path.join(image_dir, pf)).strip()

    for qn in q_nums:
        qf = normalize_filename(f"2026-06-Íµ≠Ïñ¥_{qn}.png")
        if qf not in all_files:
            print(f"‚ùå Î¨∏Ï†ú ÏóÜÏùå: {qf}")
            continue
        question = extract_text(os.path.join(image_dir, qf)).strip()

        prompt = text_prompt.format(passage=passage, question=question)
        print(f"\nüìò [Î¨∏Ï†ú {qn}]")
        try:
            print(ask_gpt(prompt))
        except Exception as e:
            print(f"‚ö†Ô∏è GPT Ïã§Ìå®: {e}")
```

```python
import os
import re
import cv2
import numpy as np
import unicodedata
import easyocr
from PIL import Image
from openai import OpenAI

# ‚úÖ Ïú†ÎãàÏΩîÎìú NFC Ï†ïÍ∑úÌôî
def normalize_filename(fn):
    return unicodedata.normalize('NFC', fn)

# ‚úÖ EasyOCR Ï¥àÍ∏∞Ìôî
reader = easyocr.Reader(['ko', 'en'], gpu=False)

# ‚úÖ ÌÖçÏä§Ìä∏ Ï∂îÏ∂ú + Î∞ëÏ§Ñ Í∞ïÏ°∞ ÏòÅÏó≠ Í∞êÏßÄ Ìï®Ïàò
def extract_text_with_underlines(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # EasyOCR Í≤∞Í≥º: (bbox, text, confidence)
    results = reader.readtext(gray, detail=True)

    # Canny + HoughÎ°ú Î∞ëÏ§Ñ ÏÑ† Í≤ÄÏ∂ú
    edges = cv2.Canny(gray, 50, 150, apertureSize=3)
    lines = cv2.HoughLinesP(edges, 1, np.pi/180, threshold=100,
                            minLineLength=50, maxLineGap=10)

    underlined_texts = set()
    if lines is not None:
        for (bbox, text, conf) in results:
            # ÌÖçÏä§Ìä∏Ïùò yÏ§ëÏã¨ Ï¢åÌëú
            text_y = np.mean([point[1] for point in bbox])
            for line in lines:
                x1, y1, x2, y2 = line[0]
                line_y = (y1 + y2) / 2
                if abs(line_y - text_y) < 10:
                    underlined_texts.add(text)
                    break

    # Ï†ÑÏ≤¥ ÌÖçÏä§Ìä∏ Ïû¨Íµ¨ÏÑ±
    full_text = ""
    for (bbox, text, conf) in results:
        if text in underlined_texts:
            full_text += f"**{text}** "
        else:
            full_text += f"{text} "

    # Î≥¥Í∏∞ Î≤àÌò∏ Îß§Ìïë
    full_text = re.sub(r'\b1[\.\)]', '‚ë†', full_text)
    full_text = re.sub(r'\b2[\.\)]', '‚ë°', full_text)
    full_text = re.sub(r'\b3[\.\)]', '‚ë¢', full_text)
    full_text = re.sub(r'\b4[\.\)]', '‚ë£', full_text)
    full_text = re.sub(r'\b5[\.\)]', '‚ë§', full_text)

    return full_text.strip()

# ‚úÖ GPT Ìò∏Ï∂ú Ìï®Ïàò
def ask_gpt(prompt_text):
    client = OpenAI()
    resp = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role":"user","content":prompt_text}],
        temperature=0.3
    )
    return resp.choices[0].message.content

# ‚úÖ GPT ÌîÑÎ°¨ÌîÑÌä∏
text_prompt = '''
Îã§ÏùåÏùÄ Íµ≠Ïñ¥ Î™®ÏùòÍ≥†ÏÇ¨ ÎπÑÎ¨∏Ìïô Î¨∏Ï†úÏûÖÎãàÎã§. ÏßÄÎ¨∏Í≥º Î¨∏Ï†úÎ•º ÏùΩÍ≥† ÏïÑÎûòÎ•º ÏàòÌñâÌïòÏÑ∏Ïöî:

1. Ïù¥ Î¨∏Ï†úÏùò Ï†ïÎãµÏùÑ ÏïåÎ†§Ï£ºÏÑ∏Ïöî (ÏÑ†ÏßÄ Î≤àÌò∏Îßå Ï∂úÎ†•: ‚ë†~‚ë§ Ï§ë ÌïòÎÇò)
2. Î∞òÎìúÏãú ÏßÄÎ¨∏ ÎÇ¥Ïö©ÏùÑ Í∑ºÍ±∞Î°ú ÌïòÏó¨ Ï†ïÎãµ ÏÑ†ÌÉù Ïù¥Ïú†Î•º ÏûêÏÑ∏Ìûà ÏÑ§Î™ÖÌï¥Ï£ºÏÑ∏Ïöî (3~5Î¨∏Ïû•)

[ÏßÄÎ¨∏]
{passage}

[Î¨∏Ï†ú]
{question}

Í≤∞Í≥ºÎäî Îã§Ïùå ÌòïÏãùÏúºÎ°ú Ï∂úÎ†•Ìï¥Ï£ºÏÑ∏Ïöî:

[Ï†ïÎãµ] ‚ë¢
[Ìï¥ÏÑ§] ‚Ä¶ (Ïó¨Í∏∞Ïóê Í∑ºÍ±∞ ÏÑ§Î™Ö)
'''

# ‚úÖ Ïù¥ÎØ∏ÏßÄ-Î¨∏Ï†ú Îß§Ìïë
mapping = {
    "p1": [1, 2, 3],
    "p2": list(range(4, 10)),
    "p3": list(range(10, 14)),
    "p4": list(range(14, 18))
}

# ‚úÖ Ïã§Ìñâ
image_dir = "/content/drive/MyDrive/classified_images/ÎπÑÎ¨∏Ìïô"
all_files = [normalize_filename(f) for f in os.listdir(image_dir)]

for p_key, q_nums in mapping.items():
    pf = normalize_filename(f"2026-06-Íµ≠Ïñ¥_{p_key}.png")
    passage = extract_text_with_underlines(os.path.join(image_dir, pf))

    for qn in q_nums:
        qf = normalize_filename(f"2026-06-Íµ≠Ïñ¥_{qn}.png")
        question = extract_text_with_underlines(os.path.join(image_dir, qf))

        prompt = text_prompt.format(passage=passage, question=question)
        print(f"\nüìò [Î¨∏Ï†ú {qn}]")
        try:
            print(ask_gpt(prompt))
        except Exception as e:
            print(f"‚ö†Ô∏è GPT Ïã§Ìå®: {e}")
```

```python
import cv2
import numpy as np
import os

def remove_underlines(input_path, output_path):
    # Ïù¥ÎØ∏ÏßÄ Î∂àÎü¨Ïò§Í∏∞
    img = cv2.imread(input_path)
    if img is None:
        raise FileNotFoundError(f"‚ùå Ïù¥ÎØ∏ÏßÄ Í≤ΩÎ°ú ÌôïÏù∏: {input_path}")

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Ïù¥ÏßÑÌôî
    _, binary = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY_INV)

    # ÏàòÌèâÏÑ† Í≤ÄÏ∂úÏö© Ïª§ÎÑê (1x30)
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (30, 1))
    detected_lines = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel, iterations=2)

    # ÏàòÌèâÏÑ† ÏòÅÏó≠ Ï†úÍ±∞
    cleaned = cv2.inpaint(img, detected_lines, inpaintRadius=3, flags=cv2.INPAINT_TELEA)

    # Ï†ÄÏû•
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    cv2.imwrite(output_path, cleaned)
    print(f"‚úÖ Î∞ëÏ§Ñ Ï†úÍ±∞ Ïù¥ÎØ∏ÏßÄ Ï†ÄÏû•Îê®: {output_path}")
```

```python
import cv2
import re
import numpy as np
import easyocr
from openai import OpenAI

reader = easyocr.Reader(['ko', 'en'], gpu=False)

def postprocess_text(text):
    corrections = {
        '@': '‚ìê',
        '1.': '‚ë†', '1)': '‚ë†',
        '2.': '‚ë°', '2)': '‚ë°',
        '3.': '‚ë¢', '3)': '‚ë¢',
        '4.': '‚ë£', '4)': '‚ë£',
        '5.': '‚ë§', '5)': '‚ë§'
    }
    for wrong, correct in corrections.items():
        text = text.replace(wrong, correct)
    return text.strip()

# Step 1: Ï†ÑÏ≤òÎ¶¨ Ïù¥ÎØ∏ÏßÄÏóêÏÑú ÌÖçÏä§Ìä∏Îßå Ï∂îÏ∂ú
def extract_clean_text(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    results = reader.readtext(gray, detail=True)
    return [(text, bbox) for bbox, text, conf in results]

# Step 2: ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÏóêÏÑú Î∞ëÏ§Ñ Í∞ïÏ°∞Îêú ÌÖçÏä§Ìä∏ Í∞êÏßÄ
def detect_underlined_texts(image_path, text_data):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 50, 150, apertureSize=3)
    lines = cv2.HoughLinesP(edges, 1, np.pi/180, 100, 50, 10)

    underlined = set()
    if lines is not None:
        for text, bbox in text_data:
            text_y = np.mean([pt[1] for pt in bbox])
            for line in lines:
                x1, y1, x2, y2 = line[0]
                line_y = (y1 + y2) / 2
                if abs(line_y - text_y) < 10:
                    underlined.add(text)
                    break
    return underlined

# Îëê Í≤∞Í≥ºÎ•º Ìï©Ï≥ê Í∞ïÏ°∞Îêú ÌÖçÏä§Ìä∏ Ìè¨Ìï®Ìïú Î¨∏ÏûêÏó¥ ÏÉùÏÑ±
def merge_text_with_underlines(text_data, underlined_texts):
    final_text = []
    for text, _ in text_data:
        if text in underlined_texts:
            final_text.append(f"**{text}**")
        else:
            final_text.append(text)
    return postprocess_text(' '.join(final_text))

# GPT Ìò∏Ï∂ú Ìï®Ïàò
def ask_gpt(prompt_text):
    client = OpenAI()
    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "user", "content": prompt_text}],
        temperature=0.3
    )
    return response.choices[0].message.content

# ÌÖúÌîåÎ¶ø
text_prompt = '''
Îã§ÏùåÏùÄ Íµ≠Ïñ¥ Î™®ÏùòÍ≥†ÏÇ¨ ÎπÑÎ¨∏Ìïô Î¨∏Ï†úÏûÖÎãàÎã§. ÏßÄÎ¨∏Í≥º Î¨∏Ï†úÎ•º ÏùΩÍ≥† ÏïÑÎûòÎ•º ÏàòÌñâÌïòÏÑ∏Ïöî:

1. Ïù¥ Î¨∏Ï†úÏùò Ï†ïÎãµÏùÑ ÏïåÎ†§Ï£ºÏÑ∏Ïöî (ÏÑ†ÏßÄ Î≤àÌò∏Îßå Ï∂úÎ†•: ‚ë†~‚ë§ Ï§ë ÌïòÎÇò)
2. Î∞òÎìúÏãú ÏßÄÎ¨∏ ÎÇ¥Ïö©ÏùÑ Í∑ºÍ±∞Î°ú ÌïòÏó¨ Ï†ïÎãµ ÏÑ†ÌÉù Ïù¥Ïú†Î•º ÏûêÏÑ∏Ìûà ÏÑ§Î™ÖÌï¥Ï£ºÏÑ∏Ïöî (3~5Î¨∏Ïû•)

‚Äª ÏßÄÎ¨∏Í≥º Î¨∏Ï†ú ÎÇ¥ÏóêÏÑú **Í∞ïÏ°∞Îêú Îã®Ïñ¥(Î≥ÑÌëú)**Îäî Î∞ëÏ§ÑÏù¥ Í∑∏Ïñ¥ÏßÑ Ï§ëÏöîÌïú ÌëúÌòÑÏù¥ÎØÄÎ°ú, ÏùòÎØ∏ ÌåêÎã®Í≥º Ï†ïÎãµ Í≤∞Ï†ï Ïãú Î∞òÎìúÏãú Ï∞∏Í≥†ÌïòÏÑ∏Ïöî.
‚ìê, ‚ìë, ‚ìí, ‚ìì, ‚ìî, „â†, <Î≥¥Í∏∞> Îì± Í∏∞Ìò∏Îì§ÏùÑ Ï†ïÌôïÌûà Ïù∏ÏãùÌïòÍ≥† Î∞òÏòÅÌïòÏÑ∏Ïöî.

[ÏßÄÎ¨∏]
{passage}

[Î¨∏Ï†ú]
{question}

Í≤∞Í≥ºÎäî Îã§Ïùå ÌòïÏãùÏúºÎ°ú Ï∂úÎ†•Ìï¥Ï£ºÏÑ∏Ïöî:

[Ï†ïÎãµ] ‚ë¢
[Ìï¥ÏÑ§] ‚Ä¶ (Ïó¨Í∏∞Ïóê Í∑ºÍ±∞ ÏÑ§Î™Ö)
'''

# Í≤ΩÎ°ú ÏßÄÏ†ï
passage_clean = "/content/drive/MyDrive/classified_images/ÎπÑÎ¨∏Ìïô/2026-06-Íµ≠Ïñ¥_p3_clean.png"
passage_raw = "/content/drive/MyDrive/classified_images/ÎπÑÎ¨∏Ìïô/2026-06-Íµ≠Ïñ¥_p3.png"

question_clean = "/content/drive/MyDrive/classified_images/ÎπÑÎ¨∏Ìïô/2026-06-Íµ≠Ïñ¥_12_clean.png"
question_raw = "/content/drive/MyDrive/classified_images/ÎπÑÎ¨∏Ìïô/2026-06-Íµ≠Ïñ¥_12.png"

# 1Îã®Í≥Ñ ÌÖçÏä§Ìä∏ Ï∂îÏ∂ú
passage_data = extract_clean_text(passage_clean)
question_data = extract_clean_text(question_clean)

# 2Îã®Í≥Ñ Í∞ïÏ°∞ Ï∂îÏ∂ú
passage_underline = detect_underlined_texts(passage_raw, passage_data)
question_underline = detect_underlined_texts(question_raw, question_data)

# Î≥ëÌï©
passage_text = merge_text_with_underlines(passage_data, passage_underline)
question_text = merge_text_with_underlines(question_data, question_underline)

# GPT ÏûÖÎ†• ÏÉùÏÑ±
prompt = text_prompt.format(passage=passage_text, question=question_text)

# Ïã§Ìñâ
print("üìò [ÏßÄÎ¨∏Í≥º Î¨∏Ï†ú OCR Í≤∞Í≥º]")
print(prompt)

print("\nü§ñ [GPT-4oÏùò ÎãµÎ≥Ä]")
try:
    print(ask_gpt(prompt))
except Exception as e:
    print(f"‚ö†Ô∏è GPT Ïã§Ìå®: {e}")
```

```python
import cv2
import re
import numpy as np
import easyocr
from openai import OpenAI

reader = easyocr.Reader(['ko', 'en'], gpu=False)

def postprocess_text(text):
    corrections = {
        '@': '‚ìê',
        '1.': '‚ë†', '1)': '‚ë†',
        '2.': '‚ë°', '2)': '‚ë°',
        '3.': '‚ë¢', '3)': '‚ë¢',
        '4.': '‚ë£', '4)': '‚ë£',
        '5.': '‚ë§', '5)': '‚ë§'
    }
    for wrong, correct in corrections.items():
        text = text.replace(wrong, correct)
    return text.strip()

# Step 1: Ï†ÑÏ≤òÎ¶¨ Ïù¥ÎØ∏ÏßÄÏóêÏÑú ÌÖçÏä§Ìä∏Îßå Ï∂îÏ∂ú
def extract_clean_text(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    results = reader.readtext(gray, detail=True)
    return [(text, bbox) for bbox, text, conf in results]

# Step 2: ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÏóêÏÑú Î∞ëÏ§Ñ Í∞ïÏ°∞Îêú ÌÖçÏä§Ìä∏ Í∞êÏßÄ
def detect_underlined_texts(image_path, text_data):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 50, 150, apertureSize=3)
    lines = cv2.HoughLinesP(edges, 1, np.pi/180, 100, 50, 10)

    underlined = set()
    if lines is not None:
        for text, bbox in text_data:
            text_y = np.mean([pt[1] for pt in bbox])
            for line in lines:
                x1, y1, x2, y2 = line[0]
                line_y = (y1 + y2) / 2
                if abs(line_y - text_y) < 10:
                    underlined.add(text)
                    break
    return underlined

# Îëê Í≤∞Í≥ºÎ•º Ìï©Ï≥ê Í∞ïÏ°∞Îêú ÌÖçÏä§Ìä∏ Ìè¨Ìï®Ìïú Î¨∏ÏûêÏó¥ ÏÉùÏÑ±
def merge_text_with_underlines(text_data, underlined_texts):
    final_text = []
    for text, _ in text_data:
        if text in underlined_texts:
            final_text.append(f"**{text}**")
        else:
            final_text.append(text)
    return postprocess_text(' '.join(final_text))

# GPT Ìò∏Ï∂ú Ìï®Ïàò
def ask_gpt(prompt_text):
    client = OpenAI()
    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[{"role": "user", "content": prompt_text}],
        temperature=0.3
    )
    return response.choices[0].message.content

# ÌÖúÌîåÎ¶ø
text_prompt = '''
Îã§ÏùåÏùÄ Íµ≠Ïñ¥ Î™®ÏùòÍ≥†ÏÇ¨ ÎπÑÎ¨∏Ìïô Î¨∏Ï†úÏûÖÎãàÎã§. ÏßÄÎ¨∏Í≥º Î¨∏Ï†úÎ•º ÏùΩÍ≥† ÏïÑÎûòÎ•º ÏàòÌñâÌïòÏÑ∏Ïöî:

1. Ïù¥ Î¨∏Ï†úÏùò Ï†ïÎãµÏùÑ ÏïåÎ†§Ï£ºÏÑ∏Ïöî (ÏÑ†ÏßÄ Î≤àÌò∏Îßå Ï∂úÎ†•: ‚ë†~‚ë§ Ï§ë ÌïòÎÇò)
2. Î∞òÎìúÏãú ÏßÄÎ¨∏ ÎÇ¥Ïö©ÏùÑ Í∑ºÍ±∞Î°ú ÌïòÏó¨ Ï†ïÎãµ ÏÑ†ÌÉù Ïù¥Ïú†Î•º ÏûêÏÑ∏Ìûà ÏÑ§Î™ÖÌï¥Ï£ºÏÑ∏Ïöî (3~5Î¨∏Ïû•)

‚Äª ÏßÄÎ¨∏Í≥º Î¨∏Ï†ú ÎÇ¥ÏóêÏÑú **Í∞ïÏ°∞Îêú Îã®Ïñ¥(Î≥ÑÌëú)**Îäî Î∞ëÏ§ÑÏù¥ Í∑∏Ïñ¥ÏßÑ Ï§ëÏöîÌïú ÌëúÌòÑÏù¥ÎØÄÎ°ú, ÏùòÎØ∏ ÌåêÎã®Í≥º Ï†ïÎãµ Í≤∞Ï†ï Ïãú Î∞òÎìúÏãú Ï∞∏Í≥†ÌïòÏÑ∏Ïöî.
‚ìê, ‚ìë, ‚ìí, ‚ìì, ‚ìî, „â†, <Î≥¥Í∏∞> Îì± Í∏∞Ìò∏Îì§ÏùÑ Ï†ïÌôïÌûà Ïù∏ÏãùÌïòÍ≥† Î∞òÏòÅÌïòÏÑ∏Ïöî.

[ÏßÄÎ¨∏]
{passage}

[Î¨∏Ï†ú]
{question}

Í≤∞Í≥ºÎäî Îã§Ïùå ÌòïÏãùÏúºÎ°ú Ï∂úÎ†•Ìï¥Ï£ºÏÑ∏Ïöî:

[Ï†ïÎãµ] ‚ë¢
[Ìï¥ÏÑ§] ‚Ä¶ (Ïó¨Í∏∞Ïóê Í∑ºÍ±∞ ÏÑ§Î™Ö)
'''

# Í≤ΩÎ°ú ÏßÄÏ†ï
passage_clean = "/content/drive/MyDrive/classified_images/ÎπÑÎ¨∏Ìïô/2026-06-Íµ≠Ïñ¥_p2_clean.png"
passage_raw = "/content/drive/MyDrive/classified_images/ÎπÑÎ¨∏Ìïô/2026-06-Íµ≠Ïñ¥_p2.png"

question_clean = "/content/drive/MyDrive/classified_images/ÎπÑÎ¨∏Ìïô/2026-06-Íµ≠Ïñ¥_9_clean.png"
question_raw = "/content/drive/MyDrive/classified_images/ÎπÑÎ¨∏Ìïô/2026-06-Íµ≠Ïñ¥_9.png"

# 1Îã®Í≥Ñ ÌÖçÏä§Ìä∏ Ï∂îÏ∂ú
passage_data = extract_clean_text(passage_clean)
question_data = extract_clean_text(question_clean)

# 2Îã®Í≥Ñ Í∞ïÏ°∞ Ï∂îÏ∂ú
passage_underline = detect_underlined_texts(passage_raw, passage_data)
question_underline = detect_underlined_texts(question_raw, question_data)

# Î≥ëÌï©
passage_text = merge_text_with_underlines(passage_data, passage_underline)
question_text = merge_text_with_underlines(question_data, question_underline)

# GPT ÏûÖÎ†• ÏÉùÏÑ±
prompt = text_prompt.format(passage=passage_text, question=question_text)

# Ïã§Ìñâ
print("üìò [ÏßÄÎ¨∏Í≥º Î¨∏Ï†ú OCR Í≤∞Í≥º]")
print(prompt)

print("\nü§ñ [GPT-4oÏùò ÎãµÎ≥Ä]")
try:
    print(ask_gpt(prompt))
except Exception as e:
    print(f"‚ö†Ô∏è GPT Ïã§Ìå®: {e}")
```
