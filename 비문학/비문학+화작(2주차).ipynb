```python
import re
import os
import shutil

# 연도와 회차 리스트
years = [2022, 2023, 2024, 2025]
exams = ['03', '06', '09', '수능']

images_dir = '/content/drive/MyDrive/output_images'
output_base = '/content/drive/MyDrive/classified_images'
nonfiction_dir = os.path.join(output_base, '비문학')
writing_dir = os.path.join(output_base, '화작')

os.makedirs(nonfiction_dir, exist_ok=True)
os.makedirs(writing_dir, exist_ok=True)

copied = False

for img in os.listdir(images_dir):
    if not img.lower().endswith('.png'):
        continue
    matched = False
    for year in years:
        for exam in exams:
            # 예: 2022-03-국어_41.png 또는 2022-03-국어_p1.png
            pattern = rf"{year}-{exam}-[\w\-]+_(p?\d+)\.png"
            m = re.search(pattern, img)
            if m:
                num_str = m.group(1)
                if num_str.startswith('p'):
                    num = int(num_str[1:])
                    # 비문학 선지 범위 예시
                    if 1 <= num <= 4:
                        shutil.copy2(os.path.join(images_dir, img), nonfiction_dir)
                        print(f"비문학 복사: {img}")
                        copied = True
                        matched = True
                        break
                    # 화작 선지 범위 예시
                    elif 9 <= num <= 11:
                        shutil.copy2(os.path.join(images_dir, img), writing_dir)
                        print(f"화작 복사: {img}")
                        copied = True
                        matched = True
                        break
                else:
                    num = int(num_str)
                    # 비문학 문제 번호 범위 예시
                    if 1 <= num <= 17:
                        shutil.copy2(os.path.join(images_dir, img), nonfiction_dir)
                        print(f"비문학 복사: {img}")
                        copied = True
                        matched = True
                        break
                    # 화작 문제 번호 범위 예시
                    elif 35 <= num <= 45:
                        shutil.copy2(os.path.join(images_dir, img), writing_dir)
                        print(f"화작 복사: {img}")
                        copied = True
                        matched = True
                        break
        if matched:
            break

if not copied:
    print("조건에 맞는 파일이 없거나, 아무 파일도 복사되지 않았습니다.")
```

```python
import os
import pytesseract
from PIL import Image

def extract_texts(images_dir):
    problem_texts = {}
    choice_texts = {}
    for img_file in os.listdir(images_dir):
        if img_file.lower().endswith('.png'):
            img_path = os.path.join(images_dir, img_file)
            try:
                # 파일명에서 _p가 있으면 선지, 없으면 문제 문항
                base = img_file.split('_')[-1]
                if base.startswith('p'):
                    choice_texts[img_file] = pytesseract.image_to_string(Image.open(img_path), lang='kor')
                else:
                    problem_texts[img_file] = pytesseract.image_to_string(Image.open(img_path), lang='kor')
            except Exception as e:
                if base.startswith('p'):
                    choice_texts[img_file] = f"OCR Error: {e}"
                else:
                    problem_texts[img_file] = f"OCR Error: {e}"
    return problem_texts, choice_texts

# 비문학
nonfiction_problems, nonfiction_choices = extract_texts(nonfiction_dir)
# 화작
writing_problems, writing_choices = extract_texts(writing_dir)

print(f"비문학 문제 개수: {len(nonfiction_problems)}")
print(f"비문학 선지 개수: {len(nonfiction_choices)}")
print(f"화작 문제 개수: {len(writing_problems)}")
print(f"화작 선지 개수: {len(writing_choices)}")

# 샘플 출력
print("\n[비문학 문제 샘플]")
for fname, text in list(nonfiction_problems.items())[:1]:
    print(f"{fname}\n{text[:500]}")

print("\n[화작 선지 샘플]")
for fname, text in list(writing_choices.items())[:1]:
    print(f"{fname}\n{text[:500]}")
```

## PDF -> 텍스트 파일로 변환

```python
import os
from PyPDF2 import PdfReader

# 연도와 시험 종류 리스트 정의
years = [2022, 2023, 2024, 2025]
exams = ['03', '06', '09', '수능']
subject = '국어'

# PDF 파일 경로 리스트 생성
pdf_paths = []
for year in years:
    for exam in exams:
        filename = f"{year}-{exam}-{subject}.pdf"
        path = f"/content/drive/MyDrive/기출/{filename}"
        pdf_paths.append(path)

# 텍스트 추출 함수
def extract_text_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    full_text = ""
    for page in reader.pages:
        text = page.extract_text()
        if text:
            full_text += text + "\n"
    return full_text

# 각 PDF 파일에 대해 텍스트 추출 및 저장
for pdf_path in pdf_paths:
    if os.path.exists(pdf_path):
        raw_text = extract_text_from_pdf(pdf_path)
        # txt 파일명 생성
        base_name = os.path.basename(pdf_path).replace('.pdf', '.txt')
        save_path = os.path.join('/content/drive/MyDrive/save_text', base_name)
        with open(save_path, 'w', encoding='utf-8') as f:
            f.write(raw_text)
        print(f"텍스트 저장 완료: {save_path}")
    else:
        print(f"파일이 존재하지 않습니다: {pdf_path}")
```

```python
import os
import re

# 저장된 텍스트 파일 경로
input_dir = '/content/drive/MyDrive/save_text'
output_path = 'questions_only.txt'

# 모든 txt 파일 목록
txt_files = [f for f in os.listdir(input_dir) if f.endswith('.txt')]

def parse_questions(text):
    # 화작 시작 위치 찾기
    hwajak_start = re.search(r'｢선택과목 \(화법과 작문\)｣', text)
    hwajak_text = text[hwajak_start.start():] if hwajak_start else ''
    nonfiction_text = text[:hwajak_start.start()] if hwajak_start else text

    # 비문학 문제 추출
    nonfiction_questions = re.split(
        r'(?:^|\n)\d+\[[^\]]+\] 다음 글을 읽고 물음에 답하시오\.|'
        r'(?:^|\n)\[\d+ ~ \d+\] 다음 글을 읽고 물음에 답하시오\.|'
        r'(?:^|\n)(?:\d+)?\[\d+\s*[~\-～]\s*\d+\]다음 글을 읽고 물음에 답하시오\.',
        nonfiction_text
    )
    nonfiction_questions = [q.strip() for q in nonfiction_questions if q.strip()]
    # 헤더 제거
    if nonfiction_questions and len(nonfiction_questions[0]) < 20:
        nonfiction_questions = nonfiction_questions[1:]

    # 화작 텍스트에서 문제 번호 앞 불필요 숫자 제거
    hwajak_text = re.sub(r'\d+(\[[^\]]+\])', r'\1', hwajak_text)

    # 화작 문제 추출 (강화 패턴)
    hwajak_pattern = r'(?:^|\n)(?:국어영역\(화법과작문\)\s*)?\[[^\]]+\].*?(?:다음은 .+?이다|\(가\)는 .+?이고,\s*\(나\)는 .+?이다)\.\s*물음에 답하시오\.'
    hwajak_questions = re.split(hwajak_pattern, hwajak_text, flags=re.DOTALL)
    hwajak_questions = [q.strip() for q in hwajak_questions if q.strip()]
    if hwajak_questions and len(hwajak_questions[0]) < 20:
        hwajak_questions = hwajak_questions[1:]

    # 각 문제에서 지문과 질문만 추출
    def extract_question(q):
        question_end = re.search(
            r'①|\n(?:국어영역\(화법과작문\)\s*)?\[[^\]]+\].*?(?:다음은 .+?이다|\(가\)는 .+?이고,\s*\(나\)는 .+?이다)\.\s*물음에 답하시오\.',
            q, flags=re.DOTALL
        )
        if question_end:
            question_text = q[:question_end.start()].strip()
        else:
            question_text = q.strip()
        return question_text

    parsed_nonfiction = [extract_question(q) for q in nonfiction_questions]
    parsed_hwajak = [extract_question(q) for q in hwajak_questions]

    return parsed_nonfiction, parsed_hwajak

with open(output_path, "w", encoding="utf-8") as f:
    for txt_file in sorted(txt_files):
        file_path = os.path.join(input_dir, txt_file)
        with open(file_path, "r", encoding="utf-8") as fin:
            text = fin.read()
        parsed_nonfiction, parsed_hwajak = parse_questions(text)
        # 비문학
        for i, q in enumerate(parsed_nonfiction, 1):
          if i <= 4:
            f.write(f"--- 비문학 문제 {i} ({txt_file}) ---\n")
            f.write(q + "\n\n")
        # 화작
        for i, q in enumerate(parsed_hwajak, 1):
          if i > 1:
            f.write(f"--- 화작 문제 {i-1} ({txt_file}) ---\n")
            f.write(q + "\n\n")
print(f"모든 txt 파일에서 문제를 추출해 {output_path}에 저장했습니다.")
```
