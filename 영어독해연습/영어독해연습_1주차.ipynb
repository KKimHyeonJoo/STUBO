## 방법(1) - partition_pdf (텍스트, 테이블, 이미지 추출)
-> 시간이 너무 오래 걸려서 다른 팀원한테 물어보기

```python
from unstructured.partition.pdf import partition_pdf

pdf_path = "/content/drive/MyDrive/Colab Notebooks/TAVE 프로젝트_STUBO/영어독해연습/data/영어독해.pdf"

elements = partition_pdf(
    filename=pdf_path,
    extract_images_in_pdf=True, # 이미지 추출(원래 True여야함)
    infer_table_structure=True, # 테이블 구조 추론
    chunking_strategy="by_title",
    max_characters=4000,
    new_after_n_chars=3800,
    combine_text_under_n_chars=2000,
    image_output_dir_path="/content/drive/MyDrive/Colab Notebooks/TAVE 프로젝트_STUBO/영어독해연습/data/pdf_images",  # 임시 저장 폴더
)

# 텍스트/테이블 분류
from pydantic import BaseModel
from typing import Any

class Element(BaseModel):
    type: str
    text: Any

categorized_elements = []
for element in elements:
    if "Table" in str(type(element)):
        categorized_elements.append(Element(type="table", text=str(element)))
    elif "CompositeElement" in str(type(element)):
        categorized_elements.append(Element(type="text", text=str(element)))

text_list = [e for e in categorized_elements if e.type == "text"]
table_list = [e for e in categorized_elements if e.type == "table"]

print(f"텍스트 개수: {len(text_list)}, 테이블 개수: {len(table_list)}")
```

## 방법(2) - pdfplumber (이미지 추출x)

```python
# 너무 속도가 느려서 이렇게...
import pdfplumber

pdf_path = "/content/drive/MyDrive/Colab Notebooks/TAVE 프로젝트_STUBO/data/영어독해.pdf"

text_list = []
table_list = []

with pdfplumber.open(pdf_path) as pdf:
    for page_number, page in enumerate(pdf.pages, start=1):
        # 페이지 텍스트
        text = page.extract_text()
        if text:
            text_list.append({"page": page_number, "text": text})

        # 페이지 테이블
        tables = page.extract_tables()
        for table in tables:
            table_list.append({"page": page_number, "table": table})

print(f"총 텍스트 페이지 수: {len(text_list)}")
print(f"총 테이블 수: {len(table_list)}")
```

- 요약 프롬프트 테스트
- 요약 모델 테스트

```python
import base64
import openai

def get_image_description(image_path):
    with open(image_path, "rb") as f:
        base64_image = base64.b64encode(f.read()).decode("utf-8")

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a tutor helping students answer English questions. Provide the correct answer with clear justification."},
            {
                "role": "user",
                "content": [
                    {
                        "type": "image_url",
                        "image_url": {
                            "url": f"data:image/png;base64,{base64_image}"
                        }
                    }
                ]
            }
        ],
        max_tokens=500,
    )

    return response.choices[0].message.content

# 이미지 해설 생성
img_summaries = []

if not image_paths:
    print("❗ PDF에서 추출된 이미지가 없습니다. 이미지 해설 생성을 건너뜁니다.")
else:
    img_summaries = []
    for p in image_paths:
        result = get_image_description(p)
        if result:
            img_summaries.append(result)

    if img_summaries:
        print("예시 이미지 해설:\n", img_summaries[0])
    else:
        print("❗ 이미지 해설 생성에 실패했습니다.")
```

## (3) vectorestore 성능 비교

### - Precision

```python
# 평가용 쿼리 및 정답 데이터 준비
eval_data = [
    {"query": "prototype 뜻이 뭐야?", "expected_answer": "원형"},
    {"query": "play down 뜻?", "expected_answer": "n ~을 축소하다, ~을 경시하다"},
    {"query": "Barbara Hillary에 관한 다음 글의 내용과 일치하지 않는 것은?", "expected_answer": "5"},
    # 실제 수능특강 문서 기반 질문으로 구성
]

# 평가 함수 정의 (Precision@1)
def evaluate_precision_at_k(retriever, eval_data, k=3):
    correct = 0
    for item in eval_data:
        query = item["query"]
        expected = item["expected_answer"]

        retrieved_docs = retriever.vectorstore.similarity_search(query, k=k)
        found = any(expected in doc.page_content for doc in retrieved_docs)

        if found:
            correct += 1

    return correct / len(eval_data)

score_gptall = evaluate_precision_at_k(retriever_gptall, eval_data, k=3)
score_llava = evaluate_precision_at_k(retriever_llava, eval_data, k=3)

print(f"GPT4AllEmbedding Precision@3: {score_gptall:.2f}")
print(f"LLaVA2Embedding Precision@3: {score_llava:.2f}")
```

### - gpt-4o-mini로 비교 평가

```python
from openai import OpenAI

client = OpenAI()

def gpt_compare(query, doc1, doc2, ground_truth):
    prompt = f"""
    [질문] {query}
    [정답] {ground_truth}

    [GPT4All 응답] {doc1}
    [LLaVA2 응답] {doc2}

    어느 응답이 정답에 더 가까운가요? 'GPT4All' 또는 'LLaVA2'로 답해주세요.
    """
    res = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
    )
    return res.choices[0].message.content.strip()

# 평가용 함수 (top 1 결과 사용 예시)
def evaluate_with_gpt(retriever_gptall, retriever_llava, eval_data):
    results = []
    for item in eval_data:
        query = item["query"]
        ground_truth = item["expected_answer"]

        doc1 = retriever_gptall.invoke(query)  # str 반환
        doc2 = retriever_llava.invoke(query)  # str 반환

        # invoke()가 str 반환하므로 그대로 사용
        doc1_text = doc1
        doc2_text = doc2

        winner = gpt_compare(query, doc1_text, doc2_text, ground_truth)
        results.append({"query": query, "winner": winner})

    return results

# 실행 예
eval_results = evaluate_with_gpt(retriever_gptall, retriever_llava, eval_data)
for r in eval_results:
    print(f"질문: {r['query']}\n우승: {r['winner']}\n")
```

```python
from langchain.chat_models import ChatOpenAI
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate

# 프롬프트 템플릿: 문제에 대한 정답 + 근거 포함
qa_prompt = ChatPromptTemplate.from_template("""
당신은 수능 영어를 전문으로 하는 한국인 튜터입니다.
아래 제공된 컨텍스트(이미지, 텍스트, 표 요약 등)를 참고하여 학생의 질문에 답하세요.
답변은 반드시 한국어로 작성하고, 정답뿐만 아니라 그 근거와 이유를 명확히 설명해주세요.

Context:
{context}

Question:
{question}
""")

# GPT-4 Turbo LLM 구성
llm = ChatOpenAI(model="gpt-4o-mini")

# RAG 체인 구성
rag_chain = (
    {"context": retriever_llava, "question": RunnablePassthrough()}
    | qa_prompt
    | llm
    | StrOutputParser()
)
```

## 방법(3) - gpt-4-vision 을 이용한 OCR (성공)

```python
import os
import re
import base64

# 정답 매핑
true_answers_dict = {
    18: 4, 19: 2, 20: 5, 21: 4, 22: 5,
    23: 5, 24: 2, 25: 4, 26: 4, 27: 2,
    28: 3, 29: 5, 30: 4, 31: 2, 32: 1,
    33: 3, 34: 4, 35: 3, 36: 3, 37: 4,
    38: 5, 39: 2, 40: 5,
    41: 3, 42: 3,
    43: 5, 44: 5, 45: 3
}

# 파일명에서 문항 번호 추출
def extract_question_numbers(filename):
    return list(map(int, re.findall(r'\d+', filename)))

# GPT 응답에서 정답 번호 추출
def extract_answer_number(text):
    patterns = [
        r'정답\s*[:：]?\s*([①②③④⑤1-5])',
        r'정답\s*([1-5])\s*번',
        r'정답은\s*([①②③④⑤1-5])\s*입니다',
        r'"([1-5])[.)]',
        r'선택\s*[:：]?\s*([1-5])',
        r'가장 적절한 선택[은는]?\s*"([1-5])"'
    ]
    for pattern in patterns:
        match = re.search(pattern, text)
        if match:
            choice = match.group(1)
            return '①②③④⑤'.index(choice) + 1 if choice in '①②③④⑤' else int(choice)
    return None

# GPT-4o OCR 함수
def gpt4o_ocr(image_path):
    with open(image_path, "rb") as f:
        base64_img = base64.b64encode(f.read()).decode("utf-8")
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "이미지에서 수능 문제 텍스트를 정확하게 추출해줘. 문항 번호가 있다면 꼭 포함해줘."},
            {"role": "user", "content": [
                {"type": "image_url", "image_url": {"url": f"data:image/jpeg;base64,{base64_img}"}}
            ]}
        ],
        max_tokens=1000
    )
    return response.choices[0].message.content.strip()

# 주 루프
for img_path in test_paths:
    filename = os.path.basename(img_path)
    question_numbers = extract_question_numbers(filename)
    if not question_numbers:
        continue

    try:
        ocr_text = gpt4o_ocr(img_path)
        selected = extract_answer_number(ocr_text)
        response_text = rag_chain.invoke(ocr_text) if selected is None else ocr_text

        for q_num in question_numbers:
            print(f"\n📘 문항 {q_num}")
            print("💡 GPT 응답:\n" + response_text.strip())
            print(f"✅ 정답: {true_answers_dict.get(q_num, '알 수 없음')}\n")

    except Exception as e:
        for q_num in question_numbers:
            print(f"\n❌ 문항 {q_num} 처리 중 오류 발생: {e}")
```
